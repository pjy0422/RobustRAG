```python
import json

for i in range(3, 10):
    with open("num_src_" + str(i) + ".json") as f:
        data = json.load(f)
    for idx, item in enumerate(data):
        item = {
            "question": item["question"],
            "correct answer": item["gt"],
            "incorrect answers": [],
            "id": idx,
            "context": [
                {
                    "id": ele["id"],
                    "text": ele["text"],
                    "score": ele["score"],
                    "hasanswer": ele["hasanswer"],
                    "title": "",
                    "link": "",
                }
                for ele in item["ctxs"]
            ],
            "expanded answer": [],
        }
        data[idx] = item
    with open("num_src_" + str(i) + ".json", "w") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)
```
